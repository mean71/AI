```python
x = np.linspace(-4, 4, 100)
y = 2**x
y1 = 2**(x-1)
y2 = 2**(x-1) + 1

plt.plot(x, y, label = '2^x')
plt.plot(x, y1, label = '2^(x-1)')
plt.plot(x, y2, label = '2^(x-1) + 1', linestyle = ":")

plt.grid(linestyle = ':')
plt.vlines(0, 0, 16, color = 'b')
plt.legend
plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = 2**-x
y1 = 2**(-x+2)

plt.plot(x, y, label = '2^-x')
plt.plot(x, y1, label = '2^(-x+2)')

plt.grid(linestyle = ':')
plt.vlines(0, 0, 16, color = 'b')
plt.legend
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
# y = 2**(x)
y_exp = np.exp(x)
y_exp2 = np.exp(-x)

# y3 = 3**(x)
# plt.plot(x, y, label = '2^x')
plt.plot(x, y_exp, label = 'e^x')
plt.plot(x, y_exp2, label = 'e^-x')

# plt.plot(x, y3, label = '3^x')

plt.grid(linestyle = ':')
plt.vlines(0, 0, 16, color = 'b')
plt.legend
plt.show()
```

```python
x = np.linspace(0, 4, 100)
y = np.log2(x)

plt.plot(x, y, label = 'log2(x)')
plt.grid(linestyle = ':')
plt.h(linestyle = ':')
plt.hlines(0, 0, 16, color = 'violet', linestyles=':')
plt.legend
plt.show()
```
