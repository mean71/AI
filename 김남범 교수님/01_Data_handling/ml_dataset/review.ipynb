```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc("font", family = "gulim")
plt.rc("axes", unicode_minus = False)
```

```python
df1 = pd.read_excel("./dataset/stock_price.xlsx")
df2 = pd.read_excel("./dataset/stock_valuation.xlsx")

print("df1 = \n",df1.head())
print("df2 = \n",df2.head())
```

```python
## concat
pd.concat([df1, df2], join = "outer", ignore_index=True)
```

```python
## join
df1.join(df2, lsuffix="x", rsuffix="y", how="inner")
df2.join(df1, lsuffix="x", rsuffix="y", how="inner")
```

```python
## merge
# pd.merge(df1, df2, on = "id", how = "left")
pd.merge(df1, df2, left_on = "stock_name", right_on = "name", how = "left")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic_group = titanic.groupby("survived")

titanic_group[["age", "fare"]].mean()
titanic_group[["age", "fare"]].std()
```

```python
titanic["age"].agg(["mean", "median", "std", "min", "max"])
titanic_group["age"].agg(["mean", "median", "std", "min", "max"])
```

```python
sns.regplot(titanic,
            x = "age",
            y = "fare",
            fit_reg = True,
            color = "r",
            order = 1,
            label = "Titanic")
plt.title("나이 VS 요금")
plt.xlabel("age")
plt.ylabel("fare")
plt.legend()
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

sns.lmplot(penguins,
           x = "bill_length_mm",
           y = "bill_depth_mm",
           col = "species",
           row = "sex",
           sharex = False,
           sharey = False)
```
