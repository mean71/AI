## Line graph(Plot) 직선 그래프

```python
# 런타임 다시 시작 (Colab 메뉴에서 "런타임" -> "런타임 다시 시작" 선택)
# 런타임 다시 시작 (Colab 메뉴에서 "런타임" -> "런타임 다시 시작" 선택)

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 나눔글꼴 경로 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font_prop = fm.FontProperties(fname=font_path, size=12)

# 데이터 생성
x = np.linspace(0, 10, 100)
y = 2 * x + 3

# 그래프 그리기
plt.figure(figsize=(10, 5))  # 그래프의 크기 설정

plt.plot(x, y, label=' 직선 그래프', color='red', linestyle='--', linewidth=2, marker='o', markersize=8)
# label: 범례 레이블, color: 선 색상, linestyle: 선 스타일, linewidth: 선 두께, marker: 마커 모양, markersize: 마커 크기

plt.title('빨간색 기다란 직선 그래프', fontproperties=font_prop, fontsize=16, fontweight='bold')
# 제목 설정, fontproperties: 한글 폰트 적용, fontsize: 폰트 크기, fontweight: 폰트 굵기

plt.xlabel('X 축(매출 원가)', fontproperties=font_prop, fontsize=14)
# X 축 라벨 설정, fontproperties: 한글 폰트 적용, fontsize: 폰트 크기

plt.ylabel('Y 축(매출 금액)', fontproperties=font_prop, fontsize=14)
# Y 축 라벨 설정, fontproperties: 한글 폰트 적용, fontsize: 폰트 크기

plt.legend(loc='upper right', prop=font_prop, fontsize=12)
# 범례 설정, loc: 범례 위치, prop: 한글 폰트 적용, fontsize: 폰트 크기

plt.grid(True, linestyle='--', alpha=0.7)
# 그리드 설정, linestyle: 그리드 스타일, alpha: 그리드 투명도

plt.annotate('점 (8, 20)', xy=(8, 20), xytext=(6, 15), arrowprops=dict(facecolor='blue', arrowstyle='->'),
             fontproperties=font_prop, fontsize=12)
# 화살표 주석 추가, xy: 주석을 추가할 점의 좌표, xytext: 주석 텍스트 위치,
# arrowprops: 화살표 스타일, fontproperties: 한글 폰트 적용, fontsize: 폰트 크기

plt.tight_layout()  # 그래프 요소가 잘리지 않도록 레이아웃 조정
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 나눔글꼴 경로 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font_prop = fm.FontProperties(fname=font_path, size=12)

# 데이터 생성
x = np.linspace(0, 10, 100)
y = 2 * x + 3

# 그래프 그리기
plt.figure(figsize=(10, 6))

plt.plot(x, y, label='직선 그래프', color='blue', linestyle='--', linewidth=2, marker='o', markersize=8)

plt.title('직선 그래프', fontproperties=font_prop, fontsize=16, fontweight='bold')
plt.xlabel('X 축', fontproperties=font_prop, fontsize=14)
plt.ylabel('Y 축', fontproperties=font_prop, fontsize=14)

plt.legend(loc='upper left', prop=font_prop, fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# 주석 추가
plt.annotate('점 (5, 13)', xy=(5, 13), xytext=(6, 15), arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontproperties=font_prop, fontsize=12, bbox=dict(boxstyle='round,pad=0.3', edgecolor='gray', facecolor='white'))

plt.tight_layout()
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 100, 100)
colors = np.random.rand(100)

# 그래프 그리기
sc = plt.scatter(x, y, s=sizes, c=colors, cmap='plasma', alpha=0.7)

plt.title('산점도', fontproperties=font_prop, fontsize=16, fontweight='bold')
plt.xlabel('X 축', fontproperties=font_prop, fontsize=14)
plt.ylabel('Y 축', fontproperties=font_prop, fontsize=14)

cbar = plt.colorbar(sc)
cbar.set_label('크기', fontproperties=font_prop, fontsize=12)
plt.tight_layout()
plt.show()
```

## Box plot

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 나눔글꼴 경로 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font_prop = fm.FontProperties(fname=font_path, size=12)

# 세 과목의 시험 점수 데이터
math_scores = [70, 80, 75, 85, 90, 78, 82, 88, 92, 95, 68, 72]
english_scores = [65, 75, 80, 88, 92, 70, 78, 85, 90, 95, 60, 68]
science_scores = [75, 85, 78, 90, 92, 80, 85, 88, 92, 98, 72, 75]

# 데이터 그룹
data = [math_scores, english_scores, science_scores]

# 그래프 그리기
plt.figure(figsize=(8, 6))
bp = plt.boxplot(data, labels=['수학', '영어', '과학'], notch=False, patch_artist=True)

# 박스 색상 설정
colors = ['pink', 'lightblue', 'lightgreen']
for box, color in zip(bp['boxes'], colors):
    box.set(facecolor=color)

plt.title('과목별 시험 점수 분포', fontproperties=font_prop, fontsize=16, fontweight='bold')
plt.xlabel('과목', fontproperties=font_prop, fontsize=14)
plt.ylabel('점수', fontproperties=font_prop, fontsize=14)

plt.xticks(fontproperties=font_prop)  # x 축 레이블에 한글 폰트 적용

plt.tight_layout()
plt.show()
```

# Seaborn

## Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 샘플 데이터 생성
flights = sns.load_dataset("flights")

# 데이터 피벗팅
data = flights.pivot(index="month", columns="year", values="passengers")

# 히트맵 그리기
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Passenger Count by Month and Year')
plt.show()
```

## Pair plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 샘플 데이터 생성
iris = sns.load_dataset("iris")

# 페어 플롯 그리기
sns.pairplot(iris, hue="species")
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

## Linear Regression plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 샘플 데이터 생성
tips = sns.load_dataset("tips")

# 선형 회귀 플롯 그리기
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.title('Linear Regression Plot of Tips')
plt.show()
```

## Bar plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 샘플 데이터 생성
tips = sns.load_dataset("tips")

# 막대 그래프 그리기
plt.figure(figsize=(8, 6))
sns.barplot(x="day", y="total_bill", data=tips, palette="Blues")
plt.title('Average Total Bill by Day')
plt.show()
```

## Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 샘플 데이터 생성
tips = sns.load_dataset("tips")

# 바이올린 플롯 그리기
plt.figure(figsize=(8, 6))
sns.violinplot(x="day", y="total_bill", data=tips, palette="Set3")
plt.title('Distribution of Total Bill by Day')
plt.show()
```

## Box plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 샘플 데이터 생성
tips = sns.load_dataset("tips")

# 박스 플롯 그리기
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title('Distribution of Total Bill by Day (Box Plot)')
plt.show()
```
