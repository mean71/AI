```python
def shuffle_pixels(img):
    # 랜덤 시드 고정
    torch.manual_seed(42)

    # (C, H, W) 형태에서 flatten
    flat_img = img.flatten()  # (C * H * W)

    # torch.randperm을 사용하여 랜덤 순서 생성
    perm = torch.randperm(flat_img.size(0))  # (C * H * W)의 랜덤 인덱스

    # 인덱스를 기반으로 텐서 값 섞기
    shuffled_img = flat_img[perm]  # 섞은 순서로 재배치

    # 원래 형상 복원
    shuffled_img = shuffled_img.view(img.shape)  # (C, H, W)

    return shuffled_img

transform = transforms.Compose([
    transforms.ToTensor(), # 이미지를 Tensor로 변환 (C, H, W)
    transforms.Lambda(shuffle_pixels), # 픽셀 섞기 함수 적용
])
```

```python
train_DS = datasets.MNIST(root = '/content/drive/MyDrive/Colab Notebooks/data', train=True, download=True, transform=transform)
test_DS = datasets.MNIST(root = '/content/drive/MyDrive/Colab Notebooks/data', train=False, download=True, transform=transform)
train_DL = torch.utils.data.DataLoader(train_DS, batch_size=BATCH_SIZE, shuffle=True)
test_DL = torch.utils.data.DataLoader(test_DS, batch_size=BATCH_SIZE, shuffle=True)
```

```python
def show_images_before_after(original, shuffled):
    # Tensor를 numpy 배열로 변환 (C, H, W -> H, W, C)
    original = original.permute(1, 2, 0).cpu().numpy()
    shuffled = shuffled.permute(1, 2, 0).cpu().numpy()

    # 두 이미지를 나란히 출력
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(original, cmap='gray')
    axes[0].set_title("Original Image")
    axes[0].axis('off')  # 축 숨기기

    axes[1].imshow(shuffled, cmap='gray')
    axes[1].set_title("Shuffled Image")
    axes[1].axis('off')  # 축 숨기기

    plt.show()

# 데이터 하나 가져오기
original_img = test_DS.data[900].unsqueeze(0)

# 섞은 이미지 생성
shuffled_img = shuffle_pixels(original_img)

# 섞기 전과 후의 이미지 비교
show_images_before_after(original_img, shuffled_img)
```

```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.fcs = nn.Sequential(nn.Linear(28*28, 30),
                                 nn.BatchNorm1d(30),
                                 nn.ReLU(),
                                 nn.Linear(30, 10))
    def forward(self,x):
        x = torch.flatten(x, start_dim=1)
        x = self.fcs(x)
        return x
```
