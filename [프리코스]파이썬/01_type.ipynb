```python
# 8월 19일 1회차
# 배운 내용
# Python 설치, VScode 설치
# Vscode 기본 사용법, Python 사용법

# 주석 작성 방법
# #(쉬프트 + 3)이 붙은 라인(줄)은 주석이 된다.

"""
큰 따옴표(") 3개를 위 아래로 연속으로 작성하면 여러 줄 주석이 된다.
큰 따옴표(") 3개 사이에 있는 코드는 모두 주석이 된다.
즉, 프로그램 작동에 영향을 주지 않는다.
"""
# VSCode 주석 단축키 : 왼쪽 컨트롤 + / (오른쪽 쉬프트 왼쪽)
# 여
# 러
# 줄
# 도
# 주
# 석
# 이
# 가
# 능
# 하다

# 파이썬(프로그래밍 언어)의 기본 자료형
# 자료형이 무엇인가? 
# 데이터의 형태 -> 자료형

# F1 레이싱 타이어 종류 -> 일반 타이어 / 레인 타이어 
# 문제를 해결하기에 적합한 자료형을 선택

# 숫자형 자료형
# 양의 정수, 음의 정수, 0(정수 자료형, int 자료형)
print(10)
print(-10)
print(0)

# 자료형을 확인하는 방법은 없을까?
# type() : 데이터(값)의 자료형을 확인하는 기능
print(type(10))
print(type(-10))
print(type(0))

# 실수형(float) 자료형
# 소수점이 있는 숫자
print(10.1)
print(-10.1)
print(0.0)
print(10.0)

# 내가 작성한 숫자가 정말 실수형 자료형일까?
# type() : 데이터의 자료형을 확인하는 기능
print(type(10.1))
print(type(-10.1))
print(type(0.0))
print(type(10.0))

# 문자열 자료형
# (문자)들의 나(열)
# 첫 번째 생성 방법 : 큰 따옴표 ""
print("'hello world'")

# 두 번째 생성 방법 : 작은 따옴표 ''
print('"hello world"')

# 문자열 사용 시 주의사항
# 숫자형과 문자열의 구분을 해야한다.
# 10을 결과로 정수형 / 문자열 만들어보자
print(10)
print("10")

# 숫자형 10과 문자열 10은 다르다 라는걸 주의해야 한다.
# type() : 데이터의 자료형을 확인하는 기능
print(type(10))
print(type("10"))

# 불린형
# 참(True) / 거짓(False) 2개의 값만 가질 수 있는 자료형
# 참(True)과 거짓(False)이 있어서 논리를 부여할 수 있다.
print(True)
print(False)

# type() : 데이터의 자료형을 확인하는 기능
print(type(True))
print(type(False))

# 형 변환 
# 자료형을 변환시키는 기능
# 예시) 정수형 -> 문자열 / 문자열 -> 정수형 / 실수형 -> 정수형
# type() 기능의 결과물 : int / float / str / bool

# 정수형(int) 변환
# int() : 다른 자료형을 정수형으로 변환하는 기능
print(int("10"))
print(type(int("10")))

# 실수형 데이터 -> 정수형 데이터로 변환
# 내림을 한다.(소수점 밑 숫자들을 버린다.)
print(int(1.5))
print(int(1.9))
print(int(1.1))
print(int(1.9999))

# 문자열 데이터 -> 정수형 데이터로 변환(오류 상황)
# print(int("10.0"))

# 숫자형(정수형, 실수형) -> 문자열 데이터로 변환
# str() : 다른 자료형 데이터를 문자열 자료형으로 변환
print(str(1.5))
print(str(0))
print(str(1))
print(str(-1))

# type() : 데이터의 자료형을 확인하는 기능
print(type(str(0)))

# 다른 자료형 데이터 -> 불린형(bool) 자료형으로 변환
# bool() : 데이터의 자료형을 불린형으로 변환하는 기능

"""
복습
불린형(bool)으로의 자료형 변환
불린형이 가지는 값 -> 참(True) / 거짓(False)
"""
# 숫자형 -> 불린형으로 변환할 때 규칙
# 숫자 0은 거짓(False) 표현
print(bool(0))

# 실수형 -> 불린형으로 변환할 때 규칙
# 실수 0.0은 거짓(False) 표현
print(bool(0.0))

# 문자열 -> 불린형으로 변환할 때 규칙
# 빈 문자열 ""은 거짓(False) 표현
print(bool(""))

# 공백(띄어쓰기)는 참? /거짓?
print(bool(" ")) # 참(True)
```
