```python
## Projector vector
def projection_vec(base_vec, proj_vec):
    scale = np.inner(base_vec, proj_vec)/np.linalg.norm(base_vec)**2 # 리니어알제브라:엘one놈 제곱하면 엘two놈
    p_vec = scale*base_vec

    plt.quiver(0, 0, base_vec[0], base_vec[1], angles = 'xy',
                scale_units = 'xy', color = 'r', scale = 1, label = 'Base vector')
    plt.quiver(0, 0, proj_vec[0], proj_vec[1], angles = 'xy',
                scale_units = 'xy', color = 'g', scale = 1, label = 'Target vector')
    plt.quiver(0, 0, p_vec[0], p_vec[1], angles = 'xy',
                scale_units = 'xy', color = 'b', scale = 1, label = 'Projected vector')
    plt.xlim(-8, 8)
    plt.ylim(-8, 8)
    plt.grid(linestyle = ':')
    plt.legend()
    plt.show()

projection_vec(np.array([-2, 1]), np.array([-6, 4]))
```
